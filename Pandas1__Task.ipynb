### 15. Change the data type of  column `'price'` from `object` to numeric

```python
df['price'] = df['price'].astype('float')
```

### 16. Show summary statistics of column `'price'`

### 17. Subset the data about cars whose `'price'` is less than `10000`

```python
filt = df['price'] < 10000
df = df.loc[filt]
```

### 18. Replace missing values in `'price'` column with the median  

```python
df[df['price'] == '?'] = df['price'].apply(pd.to_numeric, errors='coerce').median()
```

### 19. How many `'bmw'` cars are cheaper than `30000`?

```python
filt = (df['make'] == 'bmw') & (df['price'].astype('float') < 30000)
len(df.loc[filt])
```
